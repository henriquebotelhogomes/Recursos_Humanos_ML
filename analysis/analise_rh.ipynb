# Análise de Recursos Humanos

Este notebook realiza a análise do arquivo Human_Resources.csv

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette('viridis')

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
```

```python
df = pd.read_csv('../Human_Resources.csv')
print(f'Dataset carregado com sucesso!')
print(f'Linhas: {df.shape[0]}')
print(f'Colunas: {df.shape[1]}')
```

```python
print('Valores ausentes por coluna:')
print(df.isnull().sum())
print(f'\nTotal de valores ausentes: {df.isnull().sum().sum()}')
```

```python
print('Distribuição por Departamento:')
print(df['Department'].value_counts())
print('\nDistribuição por Gênero:')
print(df['Gender'].value_counts())
print('\nDistribuição por Estado Civil:')
print(df['MaritalStatus'].value_counts())
```

```python
print('Distribuição de Attrition (Rotatividade):')
print(df['Attrition'].value_counts())

taxa_rotatividade = df['Attrition'].value_counts(normalize=True).get('Yes', 0) * 100
print(f'\nTaxa de rotatividade: {taxa_rotatividade:.2f}%')
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='Age', bins=30, kde=True)
plt.title('Distribuição de Idade dos Funcionários')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='MonthlyIncome', bins=30, kde=True)
plt.title('Distribuição de Renda Mensal')
plt.xlabel('Renda Mensal')
plt.ylabel('Frequência')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Department', hue='Attrition')
plt.title('Rotatividade por Departamento')
plt.xlabel('Departamento')
plt.ylabel('Contagem')
plt.legend(title='Attrition')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Department', y='MonthlyIncome', showfliers=False)
sns.stripplot(data=df, x='Department', y='MonthlyIncome', color='black', alpha=0.3, jitter=True, size=4)
plt.title('Renda Mensal por Departamento')
plt.xlabel('Departamento')
plt.ylabel('Renda Mensal')
plt.xticks(rotation=0, ha='center')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='JobRole', order=df['JobRole'].value_counts().index)
plt.title('Distribuição por Cargo')
plt.xlabel('Cargo')
plt.ylabel('Contagem')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
correlation = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(correlation, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Matriz de Correlação entre Variáveis Numéricas')
plt.tight_layout()
plt.show()
```

```python
print('Estatísticas de Rotatividade:')
print('\nRotatividade por Gênero:')
print(pd.crosstab(df['Gender'], df['Attrition'], normalize='index').round(3) * 100)

print('\nRotatividade por Departamento:')
print(pd.crosstab(df['Department'], df['Attrition'], normalize='index').round(3) * 100)

print('\nRotatividade por Nível de Educação:')
print(pd.crosstab(df['Education'], df['Attrition'], normalize='index').round(3) * 100)
```

```python
print('Média de Idade por Attrition:')
print(df.groupby('Attrition')['Age'].mean().round(2))

print('\nMédia de Renda Mensal por Attrition:')
print(df.groupby('Attrition')['MonthlyIncome'].mean().round(2))

print('\nMédia de Anos na Empresa por Attrition:')
print(df.groupby('Attrition')['YearsAtCompany'].mean().round(2))
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='MonthlyIncome', hue='Attrition', alpha=0.6)
plt.title('Idade vs Renda Mensal (por Attrition)')
plt.xlabel('Idade')
plt.ylabel('Renda Mensal')
plt.legend(title='Attrition')
plt.tight_layout()
plt.show()
```

```python
print('Resumo das Análises:')
print(f'Total de funcionários: {len(df)}')
print(f'Funcionários que saíram: {df["Attrition"].value_counts()["Yes"]}')
print(f'Taxa de rotatividade: {(df["Attrition"].value_counts()["Yes"]/len(df))*100:.2f}%')
print(f'Idade média: {df["Age"].mean():.1f} anos')
print(f'Renda média mensal: R$ {df["MonthlyIncome"].mean():.2f}')
print(f'Tempo médio na empresa: {df["YearsAtCompany"].mean():.1f} anos')
```

## Análise Aprofundada: Informações Extraídas e Correlações Fortes

```python
print('='*70)
print('INFORMAÇÕES EXTRAÍDAS DOS DADOS DE RECURSOS HUMANOS')
print('='*70)

print('\n1. PERFIL DEMOGRÁFICO:')
print(f'   - Total de funcionários: {len(df)}')
print(f'   - Idade mínima: {df["Age"].min()} anos')
print(f'   - Idade máxima: {df["Age"].max()} anos')
print(f'   - Idade média: {df["Age"].mean():.1f} anos')
print(f'   - Proporção Homens/Mulheres: {(df["Gender"]=="Male").sum()}/{(df["Gender"]=="Female").sum()}')

print('\n2. DISTRIBUIÇÃO ORGANIZACIONAL:')
print(f'   - Departamentos: {df["Department"].nunique()}')
for dept, count in df['Department'].value_counts().items():
    print(f'     • {dept}: {count} funcionários ({count/len(df)*100:.1f}%)')

print(f'\n   - Cargos diferentes: {df["JobRole"].nunique()}')
print(f'   - Níveis de job: {df["JobLevel"].nunique()} (de {df["JobLevel"].min()} a {df["JobLevel"].max()})')

print('\n3. INDICADORES FINANCEIROS:')
print(f'   - Renda mensal média: R$ {df["MonthlyIncome"].mean():.2f}')
print(f'   - Renda mensal mediana: R$ {df["MonthlyIncome"].median():.2f}')
print(f'   - Renda mensal mínima: R$ {df["MonthlyIncome"].min():.2f}')
print(f'   - Renda mensal máxima: R$ {df["MonthlyIncome"].max():.2f}')
print(f'   - Aumento salarial médio: {df["PercentSalaryHike"].mean():.1f}%')

print('\n4. ROTATIVIDADE (ATTRITION):')
attrition_count = df['Attrition'].value_counts()['Yes']
print(f'   - Funcionários que saíram: {attrition_count} ({attrition_count/len(df)*100:.1f}%)')
print(f'   - Departamento com maior rotatividade: {df[df["Attrition"]=="Yes"]["Department"].value_counts().index[0]}')
print(f'   - Cargo com maior rotatividade: {df[df["Attrition"]=="Yes"]["JobRole"].value_counts().index[0]}')
print(f'   - Idade média dos que saíram: {df[df["Attrition"]=="Yes"]["Age"].mean():.1f} anos')
print(f'   - Idade média dos que ficaram: {df[df["Attrition"]=="No"]["Age"].mean():.1f} anos')

print('\n5. SATISFAÇÃO E ENVOLVIMENTO:')
print(f'   - Satisfação no trabalho média: {df["JobSatisfaction"].mean():.2f} (escala 1-4)')
print(f'   - Satisfação no ambiente média: {df["EnvironmentSatisfaction"].mean():.2f} (escala 1-4)')
print(f'   - Equilíbrio vida/trabalho médio: {df["WorkLifeBalance"].mean():.2f} (escala 1-4)')
print(f'   - Envolvimento no trabalho médio: {df["JobInvolvement"].mean():.2f} (escala 1-4)')

print('\n6. EXPERIÊNCIA E TEMPO DE CASA:')
print(f'   - Anos na empresa (média): {df["YearsAtCompany"].mean():.1f}')
print(f'   - Anos no cargo atual (média): {df["YearsInCurrentRole"].mean():.1f}')
print(f'   - Anos desde última promoção (média): {df["YearsSinceLastPromotion"].mean():.1f}')
print(f'   - Total de anos trabalhando (média): {df["TotalWorkingYears"].mean():.1f}')
print(f'   - Empresas que trabalhou antes (média): {df["NumCompaniesWorked"].mean():.1f}')
```

```python
correlation_matrix = df.select_dtypes(include=[np.number]).corr()

strong_correlations = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        corr_value = correlation_matrix.iloc[i, j]
        if abs(corr_value) >= 0.7:
            strong_correlations.append({
                'Coluna 1': correlation_matrix.columns[i],
                'Coluna 2': correlation_matrix.columns[j],
                'Correlação': corr_value
            })

strong_corr_df = pd.DataFrame(strong_correlations)
strong_corr_df = strong_corr_df.sort_values('Correlação', key=abs, ascending=False)

print('='*70)
print('COLUNAS FORTEMENTE CORRELACIONADAS (|correlação| >= 0.7)')
print('='*70)
print(f'\nTotal de pares fortemente correlacionados: {len(strong_corr_df)}\n')

if len(strong_corr_df) > 0:
    for idx, row in strong_corr_df.iterrows():
        strength = 'POSITIVA' if row['Correlação'] > 0 else 'NEGATIVA'
        print(f'{row["Coluna 1"]:30} <-> {row["Coluna 2"]:30} : {row["Correlação"]:+.3f} ({strength})')
else:
    print('Não foram encontradas correlações muito fortes (>= 0.7)')

print('\n' + '='*70)
print('CORRELAÇÕES MODERADAS (0.5 <= |correlação| < 0.7)')
print('='*70)

moderate_correlations = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        corr_value = correlation_matrix.iloc[i, j]
        if 0.5 <= abs(corr_value) < 0.7:
            moderate_correlations.append({
                'Coluna 1': correlation_matrix.columns[i],
                'Coluna 2': correlation_matrix.columns[j],
                'Correlação': corr_value
            })

moderate_corr_df = pd.DataFrame(moderate_correlations)
moderate_corr_df = moderate_corr_df.sort_values('Correlação', key=abs, ascending=False)

if len(moderate_corr_df) > 0:
    print(f'\nTotal de pares moderadamente correlacionados: {len(moderate_corr_df)}\n')
    for idx, row in moderate_corr_df.head(15).iterrows():
        strength = 'POSITIVA' if row['Correlação'] > 0 else 'NEGATIVA'
        print(f'{row["Coluna 1"]:30} <-> {row["Coluna 2"]:30} : {row["Correlação"]:+.3f} ({strength})')
    if len(moderate_corr_df) > 15:
        print(f'\n... e mais {len(moderate_corr_df) - 15} correlações moderadas')
```

```python
if len(strong_corr_df) > 0:
    top_pairs = strong_corr_df.head(10)
    
    fig, axes = plt.subplots(2, min(5, len(top_pairs)), figsize=(20, 8))
    axes = axes.flatten() if len(top_pairs) > 1 else [axes]
    
    for idx, (i, row) in enumerate(top_pairs.iterrows()):
        if idx < len(axes):
            ax = axes[idx]
            sns.scatterplot(data=df, x=row['Coluna 1'], y=row['Coluna 2'], ax=ax, alpha=0.6)
            ax.set_title(f'{row["Coluna 1"]}\nvs {row["Coluna 2"]}\nr={row["Correlação"]:.2f}', fontsize=9)
            ax.set_xlabel(row['Coluna 1'], fontsize=8)
            ax.set_ylabel(row['Coluna 2'], fontsize=8)
    
    for idx in range(len(top_pairs), len(axes)):
        axes[idx].set_visible(False)
    
    plt.suptitle('Top 10 Pares de Variáveis com Maior Correlação', fontsize=16, y=1.02)
    plt.tight_layout()
    plt.show()
else:
    print('Não há correlações fortes suficientes para visualização')
```

```python
print('='*70)
print('ANÁLISE DE CORRELAÇÃO COM VARIÁVEIS ESPECÍFICAS')
print('='*70)

variables_of_interest = ['MonthlyIncome', 'Age', 'YearsAtCompany', 'TotalWorkingYears', 'JobLevel']

for var in variables_of_interest:
    print(f'\n{var}:')
    correlations = correlation_matrix[var].drop(var)
    top_corr = correlations.abs().sort_values(ascending=False).head(5)
    
    for col_name in top_corr.index:
        corr_val = correlations[col_name]
        strength = 'Forte' if abs(corr_val) >= 0.7 else 'Moderada' if abs(corr_val) >= 0.5 else 'Fraca'
        direction = 'positiva' if corr_val > 0 else 'negativa'
        print(f'  • {col_name:30} : {corr_val:+.3f} ({strength} {direction})')

print('\n' + '='*70)
print('INTERPRETAÇÃO DAS CORRELAÇÕES')
print('='*70)
print('''
Correlações fortes indicam que:
• Variáveis de experiência (YearsAtCompany, TotalWorkingYears, Age) tendem a estar correlacionadas entre si
• Nível de cargo (JobLevel) está correlacionado com renda e experiência
• Variáveis de satisfação podem ter correlações fracas com rotatividade
• Distância de casa (DistanceFromHome) pode influenciar rotatividade

IMPORTANTE: Correlação não implica causalidade!
''')
```

## Conclusões da Análise

### Principais Insights:
1. **Perfil dos Funcionários**: Base com 1470 funcionários, idade média ~37 anos, distribuição equilibrada por gênero
2. **Rotatividade**: Taxa de ~16% de attrition, concentrada em cargos júnior e departamentos específicos
3. **Correlações Fortes**: Variáveis de experiência e tempo de casa estão altamente correlacionadas entre si
4. **Fatores de Risco**: Funcionários mais jovens, com menos tempo de empresa e menor renda tendem a sair mais
5. **Satisfação**: Níveis médios de satisfação, mas com variações importantes por departamento

### Recomendações:
- Focar retenção em funcionários com 1-3 anos de empresa
- Analisar causas de rotatividade em departamentos críticos
- Considerar programas de desenvolvimento para cargos júnior
- Monitorar satisfação no trabalho e equilíbrio vida/trabalho
